### Feature Engineering - Using apply function

It can be helpful for our analysis if we know how old is the movie.

We can know it using the year column and subtracting it with current year.

But we need to create a new column called movie_age which would contain this information

We will use a new function in pandas called .apply.

For using it we need to first define a function for getting the movie age as shown below

```python
def movie_age(year):
    return 2021-year
```

The below apply function applies movie_age function to every row of the dataframe movies in order to create a new column called movies_age.

We will need to use lambda function inside apply function in order to create the new column 'movie_age'

```python
movies['movie_age'] = movies.apply(lambda row:movie_age(row['year']),axis=1)
```
