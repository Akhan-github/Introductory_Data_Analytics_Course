# Dropping Columns

Usually when we get a new tabular dataset, many of the columns are not useful for further data analysis. So it is better to drop these columns in order easily deal with data as well as this reduces data size so pandas operations are executed quickly.

In the current IMDB dataset too, some columns are not necessary from the perspective of doing our project on 'Assessment of audience preference of movies and earning potential of movies.'

We will be dropping following columns<br>
* title<br>
* reviews_from_users - Number of user reviews <br>
* reviews_from_critics - Number of critic reviews<br>
* director<br>
* writer <br>
* production_company<br>
* actors<br>
* description<br>

We already have original_title column, title column is just a repetition of that.

We are dropping reviews_from_users and reviews_from_critics columns becuase avg_vote, gross income and total_votes are better measures of popularity of a movie. So we won't be considering them in our preliminary analysis.

Other columns are not useful for determining the popular movies. ( you may use director and actor column for doing the final project on your own - for now we are dropping so as to simplify further data analysis)

we use .drop function to drop columns (it can be used for rows as well), for columns we need to use the arguments 
* list of columns names to drop
* axis=1
* inplace=True

axis = 1 tells this function to drop columns (if you use axis = 0 it will drop rows)

inplace=True changes the original dataframe (if you do not use it then the dataframe won't be changed)

```python
movies.drop(['title','reviews_from_users','reviews_from_critics','director','actors','writer','production_company','description'],axis=1,inplace=True)
movies
```

# Dropping rows

If we look at the values of year column using .unique() function, we will find one of the values which does not looklike a year.

```python
movies['year'].unique()
```

Let us see what this data row looks like

```python
movies.loc[movies['year']=='TV Movie 2019']
```

This kind of corrupt data usually comes up in  your dataset, in these either you can change the dataset if you know the background information about the data or delete that data point.

In this example we will be dropping this row. We will use index of row for marking which row to delete and axis=0 to show that we want to drop rows.

```python
movies.drop([83917],axis=0,inplace=True)
```

If we now check unique values of year column, we will not find this kind of issue.

```python
movies['year'].unique()
```

# Renaming Columns

Sometimes some of the column names are incorrectly spelled.

In movies dataset you will see that worldwidegrossincome has been  incorrectly spelled

```python
movies.columns
```

It should be 'worldwide_gross_income' but it is 'worlwide_gross_income', the spelling is wrong in 'worldwide' word.

We should also rename avg_vote with imdb_score name, so as to have a clear idea of what is the imdb score.

We can rename this column correctly as shown below

```python
movies.rename({'worlwide_gross_income':'worldwide_gross_income','avg_vote':'imdb_score'},axis=1,inplace=True)
```

```python
movies
```

# Changing column datatypes

The year column has datatype object as you can see below

```python
movies.info()
```

Since it is of object datatype, we can not use it for number type operation.

For eg. we can not use this column to get all the movies released after 1950

```python
movies.loc[movies['year']<1950]
```

So we need to change the datatype of this column

We can do so using the command below

```python
movies['year'] = movies['year'].astype(int)
```

```python
movies.info()
```

Now we can use this column to get all the movies released after 1950

```python
movies.loc[movies['year']>1950]
```
