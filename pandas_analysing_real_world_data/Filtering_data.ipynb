# Filtering

## Filter column

```python
movies['language']
```

## Filter rows based on column conditions

We can also filter rows in the data with "specific values (or value ranges) within specific features."

For eg. let us suppose we want to filter movies which are made in English

**Filter movies made in english**

First we create a mask dataframe using the comparison operator '=='. 

It compares each row value of 'language' columns with 'English'.

If 'English' is in the row then it is true otherwise false.

```python
# Creating a mask
mask = movies['language']=='English'
print(mask)
```

Comparing 'language' column with above mask we can observe that wherever we have value 'English' (it should be just 'English' and not combined with other language), then it is True otherwise False

```python
movies['language']
```

Now we apply mask dataframe with movies.loc function to just select those rows where we have 'English' as the language

```python
# using mask inside .loc filters out True rows only from mask dataframe
movies.loc[mask]
```

More generally we do not initialise mask first, we write mask and filter operation in one line only

```python
movies.loc[movies['language']=='English']
```

But if you see all_lang, you will observe that many English movies are released in other languages as well.

```python
# access all the langauges in languages column
all_lang = movies['language'].unique()
all_lang[0:50]
```

The above dataframe contains those movies which are released in English only and not other languages. 

what if we want all movies which are made in English (whether or not they have other language versions)

We need to create a new mask dataframe. In this mask dataframe for each row value in langauge column we are checking if the value contains the word 'English' or not.

For applying this contains function, we need to use .str after the column.

```python
mask_str = movies['language'].str.contains('English')
print(mask_str)
```

If you observe the above dataframe you will see some of the values are NaN, these are those rows where no language value is mentioned.

If we apply the above mask operation for filtering out rows with 'English' language we will face an error

```python
# you will get an error here
movies.loc[movies['language'].str.contains('English')]
```

Mask should always have True or False value, No NA/NaN values are allowed, so we need to discard rows with NA/NaN values.

They can be discarded using the below code as shown

```python
# using na argument inside contains function to drop na rows
movies.loc[movies['language'].str.contains('English',na=False)]
```

Below we have broken down the steps that we have taken for above filtering. 

**Step-by-step explanation (from inner to outer):**<br>
* movies\['language'\] selects the language column of the data frame (technically, the object movies\['language'\] is of type pandas.Series)<br>
* movies\['language'\].str allows us to apply vectorized string methods (e.g., lower, contains) to the Series<br>
* movies\['language'\].str.contains('English') checks each element of the Series as to whether the element value has the string 'English' as a substring. The result is a Series of Booleans indicating True or False about the existence of a 'English' substring.<br>
* movies.loc\[movies\['language'\].str.contains('English')\]applies the Boolean 'mask' to the dataframe and returns a view containing appropriate records.<br>
* na = False removes NA / NaN values from consideration; otherwise a ValueError may be returned.<br>

If we just want to select the column 'original title in above filter we can do so as shown below

```python
movies.loc[movies['language'].str.contains('English',na=False),'original_title']
```

### Filter rows based on multiple column conditions

**Filter movies made in english with runtime more than 120 min**<br>
We can use & operator to check if both conditions are True then only we filter out those rows

```python
movies.loc[(movies['language'].str.contains('English',na=False)) & (movies['duration']>120)]
```
