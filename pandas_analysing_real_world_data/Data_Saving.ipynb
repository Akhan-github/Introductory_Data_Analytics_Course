# Saving the dataframe

We have learnt how to load the dataframe from a csv file in the notebook.

After some processing we may need to save the dataframe back in our folder as well

below command shows how to save the dataframe in a csv file at the following path

"Data/IMDB_rotten_tomato_dataset/IMDB/"

with file name - "Imdb_movies_saved.csv"

so final file path would be - "Data/IMDB_rotten_tomato_dataset/IMDB/Imdb_movies_saved.csv"

```python
# dataframe.to_csv(file_path,index=False) - index = False is required so that index is not saved in csv file as a column
movies.to_csv("Data/IMDB_rotten_tomato_dataset/IMDB/Imdb_movies_saved.csv", index=False)
```
