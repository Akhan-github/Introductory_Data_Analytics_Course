# Importing pandas library and data loading

```python
import pandas as pd
```

If you have saved the dataset in the previous tutorial in the file "Imdb_movies_saved.csv" file than you can access it from the following path

```python
# if you are working with this tutorial on local machine use the file path where the data is saved in your computer
movies = pd.read_csv("Data/IMDB_rotten_tomato_dataset/IMDB/Imdb_movies_saved.csv")
# We can use .head command to quickly observe the first 5 rows of the dataset
movies.head()
```

## Sorting

We can also sort our features or columns in ascending or descending order using .sort_values function. It has attributes 'ascending' which tells about the order of sorting.

```python
movies.sort_values('duration',ascending=False).head(10)
```

**Find the top 10 japanese movies with highest rating with duration less than 1 and half hours**

```python
# filter out japanese movies with duration less than 1 and half hours
japanese_movie_dur_90min = movies.loc[(movies['language'].str.contains('Japanese',na=False)) & (movies['duration']<90)]
japanese_movie_dur_90min.head()
```

```python
japanese_movie_dur_90min.sort_values('avg_vote',ascending=False).head(10)
```
