# Accessing rows and columns of dataframe - .loc and .iloc

For understanding the concept behind .loc and .iloc we will be using this dataset

```python
# dictionary of dictionary
dict_of_dict = {
    "birthyear": {"alice":1985, "bob": 1984, "charles": 1992},
    "hobby": {"alice":"Biking", "bob": "Dancing"},
    "weight": {"alice":68, "bob": 83, "charles": 112},
    "children": {"bob": 3, "charles": 0}
}
print(dict_of_dict)
```

```python
people = pd.DataFrame(dict_of_dict)
people
```

.loc is 'label based' so it can be used to access rows by using their index names and columns by their column names.

![loc.jpg](https://drive.google.com/uc?export=view&id=1unN1_pb897G9VmebrR1PhtPGWY3edZUu)

```python
# We have people dataframe with row (index) names as people names
people
```

```python
# for accessing data related to 'alice'
people.loc['alice'] # returns a Series Object
```

```python
# we can also access data of birthyear column associated with 'alice' using .loc
people.loc['alice','birthyear'] # returns a single value, in this case an integer
```

```python
# we can also access data of 2 people using loc
people.loc[['alice','bob']] # returns a dataframe Object
```

```python
# or we can just get data of column 'hobby' for these people
print(people.loc[['alice','bob'],'hobby']) # returns a Series Object OR,
print(people.loc['alice':'bob','hobby'])
```

```python
# we can just also read all the data in the column 'weight'
people.loc[:,'weight'] # returns a Series object
```

.iloc is integer index based meaning, we can access rows and columns by their integer index

![iloc.jpg](https://drive.google.com/uc?export=view&id=1qLhMTBM90GMSbSQ95zYp6hZJs4eQ4FuO)

```python
# So in people dataset
people
```

```python
# Access 1st row(alice)
people.iloc[0]
```

```python
# Access data of 1st row(alice) and 1st column(birthyear)
people.iloc[0,0]
```

```python
# Access data of first 2 rows(alice and bob)
people.iloc[0:2]
```

```python
# Access data of first 2 rows (alice and bob) for column 2nd column(hobby)
people.iloc[0:2,1]
```

```python
# Access data of 3rd column 
people.iloc[:,2]
```

Important Points regarding .loc and .iloc
* 2 values can be provided as comma separated values, first one is for row and second one is for column, though sometimes for accessing row, only one value is provided
* These values can be single value(like 'alice' or 2), a range( like 'alice':'bob' or 0:2) or a list (like [0,2]or ['birthyear'])
