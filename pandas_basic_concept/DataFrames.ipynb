# DataFrame

* DataFrame object is a 2D table, similar to a spreadsheet (with column names and row labels).
* We can combine the two series objects to form the dataframe, each of them will become a column in new DataFrame
* Data corresponding to same index belongs to same row

![dataframe.jpg](https://drive.google.com/uc?export=view&id=14WF1WQJ0bAHfnKucHGBxgDtE2TBrKQUt)

# Creating a DataFrame

In the last lesson on Series we had defined two series variables, let's define them again below and use them to create a new dataframe

```python
ser2 = pd.Series([68, 83, 112, 68], index=["alice", "bob", "charles", "darwin"],name='people')
ser2
```

```python
# Series can also be created using dictionary, The keys will be used as index labels
age = {"alice": 38, "bob": 24, "chloe": 44, "darwin": 28}
ser3 = pd.Series(age)
ser3
```

We use the command "pd.DataFrame" to create a new dataframe. (This is one of the most used command in Pandas so learn it so that you can utilise it quickly without thinking)

```python
# We can pass dictionary of Series objects to dataframe constructor
people_dict = {'weights':ser2,'age':ser3,'hobby': pd.Series(["Biking", "Dancing","Trekking"],\
                                                            index=["alice", "bob","chloe"])}
df = pd.DataFrame(people_dict)
df
```

A few things to note:
* the Series were automatically aligned based on their index
* missing values are represented as NaN
* DataFrames are nicely displayed in Jupyter Notebook

Another way to create a dataframe is by simply assigning dictionary of dictionary to dataframe constructor

```python
# dictionary of dictionary - basically inside the dictionary we have created more dictionary
dict_of_dict = {
    "birthyear": {"alice":1985, "bob": 1984, "charles": 1992},
    "hobby": {"alice":"Biking", "bob": "Dancing"},
    "weight": {"alice":68, "bob": 83, "charles": 112},
    "children": {"bob": 3, "charles": 0}
}
print(dict_of_dict)
```

```python
people = pd.DataFrame(dict_of_dict)
people
```

# Access Columns of DataFrames

You can access columns like a dictionary key as we have passed dictionary. They are returned as Series objects:

```python
df['weights']
```

You can also access multiple columns

```python
df[['weights','hobby']]
```
