# Why Series and DataFrames

The most basic element of Pandas library is Series and Dataframe.

If you recall the concept of datatypes, it was used to store variables of a particular type and it helped in applying particular type of operation to a particular data type. 

For eg. We could not divide two strings, We could use indexing operation on strings and List to quickly access the data within these variables.

In the same way Pandas has these two datatypes viz. Series and Dataframes which are used for saving and manipulating data efficiently in pandas.

So we need to first understand the concept of Series and Dataframes before we delve deeper into pandas

# Concept of Series

* A Series object is 1D array, similar to a column in a spreadsheet (with a column name and row labels).


![series](https://drive.google.com/uc?export=view&id=1DATaMUp8A_HD5iafpeaYwvT7f2_7KidR)

# Creating a Series - pd.Series

```python
# Creating Series using list
ser1 = pd.Series([1,5,6,7])
ser1
```

# Index Label
Each item in a Series object has a unique identifier called the index label. By default, it is simply the rank of the item in the Series (starting at 0) but you can also set the index labels manually:

![series_indexing.png](https://drive.google.com/uc?export=view&id=1wcNak5BaATrQ4vKAVoWfDgWGoAEuXpEg)

```python
ser2 = pd.Series([68, 83, 112, 68], index=["alice", "bob", "charles", "darwin"],name='people')
ser2
```

```python
# Series can also be created using dictionary, The keys will be used as index labels
age = {"alice": 38, "bob": 24, "chloe": 44, "darwin": 28}
ser3 = pd.Series(age)
ser3
```

# Accessing Elements in Series

```python
# we can access elements just like a dictionary
ser2['alice']
```

```python
# we can still access elements using integer location just like in regular lists
ser2[0]
```
