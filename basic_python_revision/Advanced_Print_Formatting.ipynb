# EXAMPLES - ADVANCED PRINT FORMATTING

when we introduced python print function, we learnt that it can be used to print following simple words or line or numbers or variables.

```python
# print a single word
print('hi')

# print a line
print('Python is awesome')

# print a number
print(9.6)

# print a variable
language = 'English'
print(language)
```

But print function can output more sophisticated statements as well. like
- Using variables values inside print along with a line. For eg. spoken_language = 'Hindi' and if we want to insert this variable value in print line 'My spoken language is Hindi'. In case of just 'Hindi' the value could be anything as provided by spoken_language variable.

- Print multiple variables at once along with a string line. For eg. eat_item1 = 'bread', eat_item2 = 'butter', then we can use these variables in one single print statement to output the line. 'I eat bread and butter in morning'.

We will code some of these formatting options that we can use with print function below.

# USING COMMA TO PRINT VARIABLE VALUE ALONG WITH A STRING LINE

```python
language = 'English'
# print the line "My Spoken language is English"
print("My Spoken language is",language)
```

```python
# we can take the variable value using user input and then use the above print statement code for better output
spoken_language = input('Enter Your Spoken language:') # user input
print('The spoken language of user is',spoken_language) # print statement
```

Using the above way we can also print multiple variables in one single line.

```python
marks_hindi = '80'
marks_maths = '95'
print('My marks in hindi is',marks_hindi,'and in maths is', marks_maths) # we could print multiple variables with string line
```

```python
print(marks_hindi, marks_maths) # we could just print multiple variables
```

# USING \\(BACKSLASH) OPERATOR TO PRINT QUOTES INSIDE OUTPUT LINES

You might have noticed that we always write print statement inside quotes, but if there is quote in the line that we want to print, then we may need to use \ operator.

```python
# print the following line - He said,"India's capital is delhi"
print("He said,"India's capital is delhi"")
```

```python
# printing using \ operator
print("He said,\"India\'s capital is delhi\"")
```

We can also use backslash operator to create new line (enter key functionality to create a new line) and create a tab indent.

'\n' -> new line
'\t' -> tab

let's see the uses of new line and tab character. 

```python
# using '\n' to print two sentences in different line in one single print statement
# '\n' works like a string, you can place it inside the whole string line or use it with comma operator. 
print('Python is awesome \nWe are learning print formatting options in python')

print('Python is awesome','\nWe are learning print formatting options in python')
```

```python
# using '\t' to print two sentences in different line in one single print statement
# '\t' works like a string, you can place it inside the whole string line or use it with comma operator. 
print('Python is awesome \tWe are learning print formatting options in python')


# we can also use newline and tab character in one print to get the output as shown below
print('Python is awesome','\n','\t','We are learning print formatting options in python')
```

# PRINT FORMATTING - RELATION WITH STRING FORMATTING

We had learnt in last section that if we want to print quotes inside strings as shown below then we will face syntax error

```python
print("He said,"India's capital is delhi"")
```

But the string line given above will also give an error if we assign the above line to a variable as shown below 

```python
a = "He said,"India's capital is delhi""
```

This is happening because the syntax error is not due to print function but due to string formatting problem.

As you recall we create strings by writing text inside quotes, but if quotes are present within the text then the python interpreter gets confused while reading a string text because it does not know where does the end quote for the string is coming.

To solve the above issue we also learnt that we can use backslash before each quote that needs to printed as text except the starting and ending quotes for the string as shown below. 

```python
print("He said,\"India\'s capital is delhi\"")
```

```python
a = "He said,\"India\'s capital is delhi\""
print(a)
```

Using Backslash in this manner helps python interpreter understand quotes in a text in an appropriate way. 

A backslash character in a string indicates that one or more characters that follow it should be treated specifically. (This is referred to as an escape sequence, because the backslash causes the subsequent character sequence to “escape” its usual meaning.) 


# .format METHOD FOR PRINT AND STRING FORMATTING

We have learnt above how to print multiple variables in a single print statement using comma.

.format is another function which further simplifies printing multiple variables in a single print statement. .format is actually a string function which helps in string formatting but it can be used for print formatting as well.

let's start with printing two variable with string line using .format operator to understand the syntax of .format function.

```python
item_1 = 'bread'
item_2 = 'butter'
# print the line - I eat bread and butter in morning
print('I eat {} and {} in morning'.format(item_1,item_2))
```

The above syntax can also be written as 

```python
item_1 = 'bread'
item_2 = 'butter'
# print the line - I eat bread and butter in morning
print('I eat {1} and {0} in morning'.format(item_1,item_2))
```

From above 2 codes syntax, you can understand that .format function works with a tuple of variables and then while printing it uses positions of variables inside the tuple. This relationship is shown in below image as well. 

![format_print_pic.png](https://drive.google.com/uc?export=view&id=1MN_GNfxYQ3QJWv5NmrN_AWTog8BhFguK)

We can print numbers using .format operator in similar fashion

```python
a_int = 99
b_float = 0.1
print('the value of a_int is {0} and value if b_float is {1}'.format(a_int,b_float))
```

**.format METHOD FOR FORMATTING STRINGS**<br>
Above we learnt about using .format method for print formatting options. But .format method is actually a string data type object method for positional formatting.

Let us see the use of .format method with strings.

```python
item_1 = 'bread'
item_2 = 'butter'
# store the given line in a string variable- I eat bread and butter in morning
str_line_var = 'I eat {0} and {1} in morning'.format(item_1,item_2)

print(str_line_var)
```
