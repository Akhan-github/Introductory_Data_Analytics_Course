# IMPORT LIBRARY

When we introduced python, we described that python contains good external libraries for executing code. Libraries are basically external code sources that is published for public use.

These external libraries have operations that can be performed in our code only when they are imported in our code. If we do not import python library, we can not use them in our code.

In our code below we will import two libraries that are quite useful
* math library - This library contains special math functions.
* sys library - The sys module provides information about constants, functions and methods of the Python interpreter.

Importing a library is easy we can just import a library using library name. For eg for importing math library, we can do importing using import math

```python
# import python math library
import math
# now we can performa special functions inside math library - for eg print the value of pi or find the cos of an angle
print(math.pi)
print(math.cos(0))
```

```python
# import sys library
import sys
# we can use the sys library to print the folder path of the places from where python interpreter is working
print(sys.path)
```

# IMPORTING MODULES

When our program grows bigger, it is a good idea to break it into different modules.

A module is a file containing Python statements. Python modules have a filename and end with the extension .py.

Definitions inside a module can be imported to another module or the jupyter notebook in Python. We use the import keyword to do this.

Let's try to create and import a module file called 'python_math_constant.py' in our current notebook.
