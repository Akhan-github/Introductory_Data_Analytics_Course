# For loops

A for loop can iterate over a collection of values (lists, strings, dictionaries, etc.) The indented code is executed for each item in the collection of values.

```python
# For loop run over a list
cities = ["delhi", "kolkata", "mumbai"]
for city in cities:
    print (city)
```

```python
# For loop run over a string - it will print each character of the string
city_name = 'Delhi'
for city_name_char in city_name:
  print(city_name_char)
```

We can use range function with for loops to iterate over sequence of integers such as shown below. 

```python
for i in range(4): # range(4) => sequence of integers from 0 to 4-1 or 3
    print(i)
```

**Summing using for loops**<br>

Let's suppose we want to sum all the elements in the following numbered list

```python
list_num = [90,10,30,50,60]

# write the code using for loop and sum variable to get the sum of all numbers in above list.
sum_list_num = 0

for num in list_num:
    print(num)
    sum_list_num = sum_list_num+num
    
print(sum_list_num)
```

**Iterating through a dictionary**

We can also iterate through a dictionary using for loop for accessing each key and value.

We need to use .items method of dictionary to get both key and value in dictionary.

let's just first look into .items methods of dictionary

```python
student_data = {'Name':'Raj','Age':15,'Sex':'M','Class':10,'Grade':'A'}

print(student_data)
print(student_data.items())
print(type(student_data.items()))
```

So applying .items() method over dictionary simply returns a dict_items data type. dict_items is a list of tuples. each tuple is key,value pair. And the order of key , value in dict_items is same as the original dictionary that we built.

We can access the key, value of dictionary using this dict_items created by .items() method as shown below

```python
for key_val,value_val in student_data.items(): # at each iteration we get a tuple of key,value pair 
    print(str(key_val)+'-->'+str(value_val))   # the output of key, value is saved in variable key_val, value_val
```

So you can see for loop can also be used to access elements of a dictionary.

**Break and Continue**

When the loop encounters the break command, the loop will terminate immediately. If there were more items in the list, they will not be processed.



```python
# `break` from a for loop
cities = ["delhi", "kolkata", "mumbai"]
for city in cities:
    if city == "kolkata":
        break
    print (city)
```

When the loop encounters the continue command, the loop will skip all other operations for that item in the list only. If there were more items in the list, the loop will continue normally.

```python
# `continue` to the next iteration - kolkata will not be printed, loop will continue to mumbai
cities = ["delhi", "kolkata", "mumbai"]
for city in cities:
    if city == "kolkata":
        continue
    print (city)
```
