# Dictionary

A dictionary consists of a collection of key-value pairs i.e. a single item in dictionary is key-value pair.

Each key-value pair maps the key to its associated value.

Each key is separated from its value by a colon (:), the items are separated by commas, and the whole thing is enclosed in curly braces.

The following code block defines a dictionary that maps student data entry keys to corresponding value.



```python
student_data = {'Name':'Raj','Age':15,'Sex':'M','Class':10,'Grade':'A'}

print(student_data)
```

# Accessing items in a dictionary

The entries in the dictionary display in the order they were defined. 

```python
student_data = {'Name':'Raj','Age':15,'Sex':'M','Class':10,'Grade':'A'}

print(student_data) # First print
```

```python
print(student_data) # Second print
```

But display order is irrelevant when it comes to retrieving them. Dictionary elements are not accessed by numerical index as we have seen in the case of lists. This is also the reason that dictionary is considered an unordered collection.

so below code will fail

```python
student_data[1]
```

The above code says KeyError because it expects that 1 is a key of dictionary and now it is trying to access the value associated with key 1 but during execution it sees that key 1 does not exist so it says KeyError.

We can access any value associated to key in case of dictionary rather than indices.

Keys can be used either inside square brackets \[\] or with the get() method.

If we use the square brackets \[\], KeyError is raised in case a key is not found in the dictionary.

```python
print(student_data['Name']) # print the value associated with key 'Name' i.e. 'Raj'

print(student_data['Grade']) # prints the value associated with key 'Grade' i.e. 'A'

print(student_data['School']) # gives KeyError as there is no key 'School' in dictionary.
```

We can also access values associated with a key using get method of dictionary as shown below. the get() method returns None if the key is not found.

```python
print(student_data.get('Name')) # print the value associated with key 'Name' i.e. 'Raj'

print(student_data.get('Grade')) # prints the value associated with key 'Grade' i.e. 'A'

print(student_data.get('School')) # gives None as the output as there is no key 'School' in dictionary. - no error
```
