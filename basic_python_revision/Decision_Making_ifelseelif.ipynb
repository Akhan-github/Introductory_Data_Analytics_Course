## Boolean Datatype and Operators

Before learning if else conditionals it is better to understand what boolean dataype and operators which work on it as this forms the backbone of working with if else conditional statements

**Boolean Datatype**<br>
Boolean is a logical data type that represents two values either True or False.

True and False are actually used as keywords in Python for Boolean data type. We do not type quotes around True and False values, otherwise they would be treated as strings.

Note the keywords True and False must have an Upper Case first letter. Using a lowercase true returns an error.

We can represent boolean variable True or False values in the following way



```python
# assign True value to Bool_var_true - there is no quotes written around True while assigning it to the variable
# If you put quotes around True then 'True' will become a string
Bool_var_true = True
Bool_var_false = False
print(Bool_var_true)
print(Bool_var_false)
```

**Comparison Operators**<br>
Comparison operators are used to compare values. It returns either True or False according to the condition.

For eg. let's suppose we use greater than comparison operator (>) on numbers.<br>
5>3<br>
the above statement is True, the above statement is actually evaluated as a boolean

```python
5>3
```

The below example shows the result of applying comparison operators to numeric datatypes

```python
# equal to and not equal to operator works with both string and numbers
a = 200
b = 250
c = 250
d = 250.0
# using '==' and '!='  with numbers
print(a == b) # False

print(a != b) # True

print(b == c) # True

print(b != c) # False

# comparing a floating variable with integer variable using '==' with same value will give True
print(c == d)
```

The below example shows the result of applying comparison operators to strings

```python
str_a = 'Python'
str_b = 'python'
str_c = 'python'

# using '==' and '!='  with strings
print(str_a == str_b) # False

print(str_a != str_b) # True

print(str_b == str_c) # True

print(str_b != str_c) # False
```

**Logical Operators**<br>
Python has three logical operators 'and', 'or' and 'not'. These operators work on boolean variables.

In case of comparison operator we saw that we operate comparsion operators on numbers and in some cases strings and the final output is always a boolean value.

In case of Logical operators it always works on boolean variables and the resulting value is also always a boolean value. For eg. In the expression - A and B - A and B are boolean variables and 'and' is a logical operator.

The three logical operators are defined as below

'and' - True if both the operands are true<br>
'or' - True if either of the operands is true<br>
'not' - True if operand is false (complements the operand)


```python
# let's suppose we have three boolean variables
A = 5>3 #will evaluate to True
print(A)
B = 4!=4 # will evaluate to False
print(B) 
C = False 
print(C)
```

```python
# Using 'or' operator
print(A or B) # True

# we can also assign the value of logical operator to a variable
result = A or B
print(result)
```

```python
# Using 'and' operator
print(A and B) # False
```

```python
# Using 'not' operator
print(not A) # False

print(not B) # True
```

## if else conditional statements

We can use if statements to conditionally do something. The conditions are defined by the words if, elif (which stands for else if) and else. We can have as many elif statements as we want. The indented code below each condition is the code that will execute if the condition is True.

```python
# If statement
x = 4
if x < 1:
    score = 'low'
elif x <= 4: # elif = else if
    score = 'medium'
else:
    score = 'high'
print (score)
```

```python
# If statement with a boolean
x = True
if x:
    print ("it worked")
```

We can also use if else statement to check if an element is in a list or not. 

We need to use 'in' operator to use if else with lists

```python
cities = ["delhi", "kolkata", "mumbai"]

if 'delhi' in cities:
  print('delhi is in the list')
```
