# Functions - Introduction and Why?

**Definition and why we use functions**<br>
Functions are a way to modularize reusable pieces of code i.e. once a piece of code is written as function we can use it any number of times anywhere in our code which reduces our effort.

They're defined by the keyword def which stands for definition and they can have the following components.

Let' see an example of using a function in code

```python
def print_lines():                             # FUNCTION HEADER
    """Function for printing a string"""       # DOC STRINGS
    print("Python is awesome")                 # FUNCTION BODY STATEMENT
    print("This is Python Function Tutorial")  # FUNCTION BODY STATEMENTS
```

Let's see how we can call the function to execute the code inside the function

```python
print_lines()                                  # FUNCTION CALL LINE
print("function call is over")                 # FUNCTION CALL IS OVER
```

# Function Arguments

Information can be passed into functions as arguments. Arguments are also referred to as parameters.

Arguments are specified after the function name, inside the parentheses.

we can create a new function 'print_name' by giving it the parameter as name variable which contains the name of a person

let's look into how the function print_name looks like.

```python
def print_name(name): # name is a variable which is being used as argument for the function
    """Print the name of person"""
    print('The name of person is',name)
    
print_name('JUDY')
```

So in the above function, when we gave string 'JUDY' inside print_name function while calling it.

'JUDY' was assigned to 'name' variable used in function print_name. Now where ever we use name variable inside function, the value of name variable 'JUDY' will be used.

**Multiple Arguments**<br>
We can also pass multiple arguments to a function. We need to separate each of them by a comma.

let's see a function which takes length and breadth of a rectangle and computes its area.



```python
def compute_rectangle_area(length, breadth):
    """Compute area of rectangle"""
    area = length * breadth # using variable length and breadth
    print('Area of Rectangle with length',length,'and breadth',breadth,'is',area)
l = 5
b = 4    
compute_rectangle_area(l,b)
```

In the above code value inside variable 'l' which is 5 is passed as value to variable 'length' and value inside variable 'b' which is 4 is passed as value to variable 'breadth'.

The order of variable is important in this case, so first value is passed to first variable and second value to second variable.



# Function return statement

In our program to compute the area of rectangle, we are finding the area inside the function and then printing the area from inside the function only.

however it is better to just find the area inside the function and print the result somewhere else.

let's see how we can do that using return statement

```python
def compute_rectangle_area(length, breadth):
    """Compute area of rectangle"""
    area = length * breadth # using variable length and breadth
    return area             # in place of print statement we have used return keyword to output area variable value outside
                            # the function
    
l = 5
b = 4

area_rectangle = compute_rectangle_area(l,b) # the value returned from function should be assigned to a new variable


# Now we can use the variable area_rectangle, l and b to print the statement
print('Area of Rectangle with length',l,'and breadth',b,'is',area_rectangle)
```

In the above code, after we compute area variable inside function, we pass it to return statement.

returns statements basically shifts the program back to the function call i.e. we have exited the function.

Other than exiting, return also carries the value of area variable which it assigns to variable area_rectangle.
