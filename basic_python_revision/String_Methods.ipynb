# String Methods

As we saw in the above case, when we use for loop over a string it prints the characters inside the string one by one.

This proves string is a collection of character just like list is collection of some objects

So we can use indexing methods as we learnt in list in the case of strings as well

```python
city_name = 'Delhi'

print(city_name[0]) #prints D
print(city_name[2]) #prints l
```

Since we can apply indexing operations over strings, we can use the same indexing operations as we have learnt in lists. For eg. we can use the negative indexing as shown in the lists

```python
print(city_name[-1]) #prints i
```

In the same way we can use slicing operations over strings and also reversing operation that we learnt in list in the same way in strings. For eg. we can reverse a string as shown below

```python
print(city_name[::-1])
```

There are some useful function that would be great to know if we are working with strings. Some of these methods will be used when we are working on our project<br>
* 'in' operator - to check if a character or substring is in string or not
* join() - joins all the elements in a sequence object like list or tuple to form a string
* split() - splits a string based on given splitting character into a list
* replace() - replace a substring in a string with a given string.
* strip() - removes leading and trailing spaces in a string

Below we are showing how each of these methods are used.

```python
# We can test if a substring exists within a string or not, using the keyword in. 
# The expression with in keyword always returns a boolean value.
exp = 'press' in 'expression'
print(exp)
```

in is used most of the times with if else code block for checking if a substring exist in a string

```python
str_list = ['mariner','martini','matrimony','mars','march']
            
# the below for loop prints all the string in the list which contains the substring 'mar'
for str_text in str_list:
    if 'mar' in str_text:
        print(str_text)
        
```

```python
# using join method
str_list = ['This','is','a','string','line']

# we are using space to join all the elements in above list to print the resulting string
print(' '.join(str_list))
```

```python
# using split method
str_line_1 = 'C:/Desktop/Python'

print(str_line_1.split('/'))

str_line_2 = 'This is a string line'

print(str_line_2.split(' '))
```

```python
# using replace method
str_text = 'walker'

# we are replacing 'walk' word with 'mark' word
print(str_text.replace('walk','mark'))
```

```python
# using strip method
str_text = '  walker   '
print(str_text) # has spaces at the start and at the end

print(str_text.strip()) # no spaces
```

The methods will be quite useful when we start working on data analysis in Pandas
