# PYTHON VARIABLES

A variable is a location in memory used to store some data (value).

They are given unique names to differentiate between different memory locations. 

We don't need to declare a variable before using it. In Python, we simply assign a value to a variable and it will exist. 

We don't even have to declare the type of the variable. This is handled internally according to the type of value we assign to the variable.

```python
#We use the assignment operator (=) to assign values to a variable
# int variable and print the variable
a_int = 9
print(a_int)
```

```python
# float variable and print the variable
b_float = 10.23
print(b_float)
```

```python
# string variable and print the variable
language = 'English'
print(language)
```

String variables need not be a single word, it can be a sentence as well as shown below. The only requirement is that the value should be inside quotes.

```python
# string variable which is a sentence
lang_line = 'I can speak English'
print(lang_line)
```

we can change the value of variable to a new value from an old value in the following manner.

```python
# change variable values 
print(language)
language = 'Hindi'
print(language)
```

```python
# change variable value using another variable using strings
print(language)
official_language = 'English'
language = official_language
print(language)
```

```python
# change variable value using another variable using numbers
print(a_int)
b_int = 10
a_int = b_int
print(a_int)
```

We can also access memory location of a variable

```python
# use print(id(var_name)) for printing memory location
print(id(a_int))
```

We talked about the type of variables above, the type of variable is also called 'DATA TYPE'. Every variable in python has data type. 

In this section we learnt about three types of data types:
* String
* Integer
* Floats

**VARIABLE NAMING RULES/SUGGESTIONS**
* variable names should have a combination of letters in lowercase (a to z) or uppercase (A to Z) or digits (0 to 9) or an underscore (_). For example: spoken_language, FIRST_NAME, number, CaptialCity

* We should never use 'space' in variable name. For eg. 'spoken langauge' is wrong variable name.

* variable names can never start with numbers. For eg. 1number is wrong variable name.

* Connot use certain words as variable names. These particular words are also called Keywords. For eg. you cannnot use 'else' as  variable name.

* We should use good descriptive variable names such that once we read the code we must be able to understand what the variable signifies. For eg. we used 'langauge' as variable name. We could have just used 'l' as variable name but it would not tell us what this variable signifies.

* To make variable names descriptive we may need to use variable name with more than one word. we can write it using underscores. Foreg. we wrote the variable name for spoken language as 'spoken_language'.
