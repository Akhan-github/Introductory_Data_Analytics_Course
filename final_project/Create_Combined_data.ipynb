# **Import pandas and loading cleaned IMDB data and rotten tomato movies data**

```python
import pandas as pd
```

We will load the 

1. movies_cleaned.csv file (this is the cleaned IMDB data file)

2. rotten_tomato_movies.csv file (this is the rotten tomato file with movies and tomatometer rating data)



```python
imdb_data = pd.read_csv("Data/IMDB_rotten_tomato_dataset/IMDB/movies_cleaned.csv")

rotten_tomato_data = pd.read_csv("Data/IMDB_rotten_tomato_dataset/rotten_tomatoes/rotten_tomatoes_movies.csv")
```

In this lesson we will clean up rotten tomato dataset and then finally combine it with IMDB dataset. The final combined dataset will be used for our final project analysis.

# **Cleaning the rotten tomato dataset**

```python
rotten_tomato_data.head(10)
```

```python
rotten_tomato_data.info()
```

Select movie_title, original_release_date and tomatometer_rating column

```python
rotten_tomato_data = rotten_tomato_data[['movie_title','original_release_date','tomatometer_rating']]
```

```python
rotten_tomato_data
```

Remove rows which has null values for original_release_date

```python
rotten_tomato_data = rotten_tomato_data[rotten_tomato_data['original_release_date'].notnull()].copy()
```

Create a year column out original_release_date column and then drop original_release_date column

```python
def get_year(release_date):
  year = release_date.split('-')[0]
  return year

rotten_tomato_data['year'] = rotten_tomato_data.apply(lambda row:get_year(row['original_release_date']),axis=1)
```

```python
rotten_tomato_data.drop(['original_release_date'],axis=1,inplace=True)

rotten_tomato_data
```

# **Combining the dataset**

```python
imdb_data
```

Renaming movie_title column to original_title in rotten_tomato_data dataframe so as to join on same column name

Also convert the datatype of year column

```python
rotten_tomato_data.rename({'movie_title':'original_title'},axis=1,inplace=True)

rotten_tomato_data
```

```python
rotten_tomato_data['year'] = rotten_tomato_data['year'].astype(int)
```

We will be joining on two columns viz. original_title and year.

Joining just on original_title column leads to wrong type of join because there are several movies with same names released in multiple years. So for these movies the data is not joined properly.

```python
combined_data = pd.merge(imdb_data,rotten_tomato_data,on=['original_title','year'])

combined_data
```

In the joined dataframe we have got the tomatometer rating column at the last from rotten_tomato_data.

# **Cleaning up the combined data**

Clean up the unnecessary columns

```python
combined_data.drop(['imdb_title_id','date_published','votes','budget','usa_gross_income',\
               'metascore','movie_age'],axis=1,inplace=True)

combined_data
```

Cleaning up the worldwide_gross_income column

```python
combined_data = combined_data.loc[combined_data['worldwide_gross_income'].str.startswith('$',na=False)].copy()
```

```python
combined_data.info()
```

```python
def remove_dollar_sign(income):
  return int(income.replace('$','').strip())
```

```python
combined_data['worldwide_gross_income'] = combined_data.apply(lambda row: remove_dollar_sign(row['worldwide_gross_income']),axis=1)
```

```python
combined_data['worldwide_gross_income'] = combined_data['worldwide_gross_income'].astype(int)
```

```python
combined_data.info()
```

We need to scale worldwide_gross_income by dividing it with 1000000 so that we can work with this column easily.

We also need to scale imdb_score column by multiplying it with 10 so that we can compare the IMDB ratings with tomatometer ratings easily.

```python
combined_data['worldwide_gross_income'] = combined_data['worldwide_gross_income']/1000000
combined_data['imdb_score'] = combined_data['imdb_score']*10
```

```python
combined_data
```

# **Saving the combined data**

We will save the final combined file in the csv file called 'combined_final_data.csv'

```python
combined_data.to_csv('Data/IMDB_rotten_tomato_dataset/combined_final_data.csv',index=False)
```
